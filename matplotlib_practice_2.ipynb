```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, "o")
#creates a scatter plot just by adding the parameter o inside the parentheses (tells it to switch from line to scatter)
```

```python
plt.scatter(x_axis, y_axis)
#creates a scatter plot using the scatter plot function
```

```python
plt.scatter(y_axis, x_axis, color="red", label= "Chicago")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
#S reflects the marker size, so setting it to y axis will set the marker size to increase with y
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
#triples the size of the markers using the y axis list as a base
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
#refactored to increase bubbles more simply
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
#creates scatter plot in object oriented interface
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", alpha=.20, linewidths=2, label="Boston", edgecolor="black")
plt.legend()
plt.gca().invert_yaxis()
plt.xlim(0, 50)

#alpha sets the transparency level and must be between 0 and 1 (.2 is 20% transparent)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
#increases the size of the chart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
plt.legend(labelcolor="white")
plt.show()
```

```python
#create a pie chart using the object oriented interface
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)

plt.show()
```
